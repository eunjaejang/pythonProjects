## BDP, BDH, BDS 속성들 처리 및 데이터 유형
- securities - (type) String 
- startDate, endDate - (type) datetime or '%Y%m%d' (string)
- OVERRIDES - (type) Dict ex) {END_DATE_OVERRIDE:datetime(2017,2,17)}
- SubField - (kwargs) ex) abdc='####', bdcd='###' ~ in sequence

```python
bloomberg.bdp(['IBM US Equity', 'AAPL US Equity'],'name')
#다수의 종목과 필드 입력가능
```

```python
bloomberg.bdp(['IBM US Equity', 'AAPL US Equity'],'BEST_EPS',{'BEST_FPERIOD_OVERRIDE':'1FQ'})
#Override처리 예시
```

```python
bloomberg.bdp(['IBM US Equity', 'AAPL US Equity'],'BEST_EPS',{'BEST_FPERIOD_OVERRIDE':'2FQ'})
#Override처리 예시
```

```python
bloomberg.bdp('US900123AL40 Govt', 'YLD_YTM_BID', {'PX_BID':'153.1250'})
#Override처리 예시
```

```python
bloomberg.bdh('AAPL US Equity', 'PX_LAST', '20141231', '20150110')
#Historical데이터 예시
```

```python
bloomberg.bdh('AAPL US Equity', 'PX_LAST', '20141231', '20150110', currency="KRW")
#Historical에 kwargs 사용예시
```

```python
bloomberg.bdh('AAPL US Equity', 'PCT_CHG_INSIDER_HOLDINGS', '20141231', '20150131', periodicitySelection='WEEKLY')
#Historical에 kwargs 사용예시2
```

```python
bloomberg.bds('MXKR Index','INDX_MWEIGHT')
#BulkData 예시
```

```python
bloomberg.bds('MXKR Index','INDX_MWEIGHT_HIST',{'end_date_override':datetime(2017,2,17)})
#BulkData 에서 override 예시
```

```python
bloomberg.bds(['MXKR Index', 'MXCN Index'],'INDX_MWEIGHT')
#두개 이상 Index의 BulkData 예시
```
